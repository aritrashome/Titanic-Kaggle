```python
import numpy as np
import matplotlib.pyplot as plt  #Learn Dot plot and tree map and countplot also
import pandas as pd
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
#plt.scatter(df['Embarked'],df.Fare)
```

```python
#Preprocessing
y = df.Survived
X = df.drop(['Survived','PassengerId','Name','Ticket','Cabin'],axis=1)


X['Age'].fillna(X['Age'].mean(), inplace=True)

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
lab_enc_x = LabelEncoder()
X['Sex'] = lab_enc_x.fit_transform(X['Sex'])

X = pd.concat([X, pd.get_dummies(X['Embarked'])], axis=1)
X = X.drop(['Embarked',],axis=1)
```

```python
#Rough  DONT COMPILE
bokka =  X.Cabin.isnull()
for i in range(len(X)):
    if bokka[i] == False:
        X.Cabin[i] = 1
    else: X.Cabin[i] = 0
```

```python
X.head(20)
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
y_df = pd.DataFrame()
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression().fit(X_train,y_train)
y_df['Logistic'] = clf.predict(X_test)
y_pred2 = clf.predict(X_train)

acc2=accuracy_score(y_train, y_pred2)
acc1=accuracy_score(y_test, y_df['Logistic'])
print(acc2, acc1)
```

```python
#Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(X_train,y_train)
y_df['DecTree'] = classifier.predict(X_test)
y_pred2 = classifier.predict(X_train)

acc2=accuracy_score(y_train, y_pred2)
acc1=accuracy_score(y_test, y_df['DecTree'])
print(acc2, acc1)
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=1)
classifier.fit(X_train,y_train)
y_df['RanFor'] = classifier.predict(X_test)
y_pred2 = classifier.predict(X_train)

acc2=accuracy_score(y_train, y_pred2)
acc1=accuracy_score(y_test, y_df['RanFor'])
print(acc2, acc1)
```

```python
#SVM Classifier
from sklearn.svm import SVC
classifier = SVC(kernel='rbf')
classifier.fit(X_train,y_train)
y_df['SVM'] = classifier.predict(X_test)
y_pred2 = classifier.predict(X_train)

acc2=accuracy_score(y_train, y_pred2)
acc1=accuracy_score(y_test, y_df['SVM'])
print(acc2, acc1)
```

```python
#Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
y_df['Naive'] = classifier.predict(X_test)
y_pred2 = classifier.predict(X_train)

acc2=accuracy_score(y_train, y_pred2)
acc1=accuracy_score(y_test, y_df['Naive'])
print(acc2, acc1)
```

```python
#XGBoost
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train,y_train)
y_df['XGB'] = classifier.predict(X_test)
y_pred2 = classifier.predict(X_train)

acc2=accuracy_score(y_train, y_pred2)
acc1=accuracy_score(y_test, y_df['XGB'])
print(acc2, acc1)
```

```python
#KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=9,metric='minkowski',p=2)
classifier.fit(X_train,y_train)
y_df['KNN'] = classifier.predict(X_test)
y_pred2 = classifier.predict(X_train)

acc2=accuracy_score(y_train, y_pred2)
acc1=accuracy_score(y_test, y_df['KNN'])
print(acc2, acc1)
```
